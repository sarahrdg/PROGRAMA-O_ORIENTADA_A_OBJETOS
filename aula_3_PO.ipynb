```python
import math

def calcular_distancia(p1,p2):
  return math.sqrt((p2[0] - p1[0])**2 + (p2[1] - p1[1])**2)
#calcula a distância entre dois pontos

def coordenada(indice,nome_ponto):
  while True:
    try:
      x = float(input(f'Digite a coordenada x do {nome_ponto} {indice}: '))
      y = float(input(f'Digite a coordenada y do {nome_ponto} {indice}: '))
      #print('-'*20)
      return (x,y)
    except ValueError:
      print('Entrada inválida. Por favor, digite um número!')
#solicita e valida que a entrada dos valores, garantindo que sejam números

def triangulo(a,b,c):
  #arredonda os lados para 6 casas decimais (paracorrrigir o erro do float) e ordena do menor -> maior.
  lados = sorted([round(a,6) , round(b,6) , round(c,6)])

#se a soma entre os dois ados for menor q o terceiro, não é um triângulo.
  if lados[0] + lados[1] <= lados[2]:
    print('Não é um triângulo')
    return

  if lados[0] == lados[2]:
    print('Triângulo equilátero, pois possui os três lados de mesma medida.')
  elif lados[0] == lados[1] or lados[1] == lados[2]:
    print('triângulo isósceles')
  else:
    print('Triângulo escaleno')
  #classifica o triângulo baseado no comprimento dos 3 lados

def quadrilatero(pontos):
    lados = [
        calcular_distancia(pontos[0], pontos[1]), # Lado 1-2
        calcular_distancia(pontos[1], pontos[2]), # Lado 2-3
        calcular_distancia(pontos[2], pontos[3]), # Lado 3-4
        calcular_distancia(pontos[3], pontos[0])  # Lado 4-1
    ]
    diagonais = [
        calcular_distancia(pontos[0], pontos[2]), # Diagonal 1-3
        calcular_distancia(pontos[1], pontos[3])  # Diagonal 2-4
    ]
#arredonda os valores para comparação segura
    l = [round(l,6) for l in lados]
    d = [round(d,6) for d in diagonais]

#define as regras
    lados_iguais = (l[0] == l[1] and l[1] == l[2] and l[2] == l[3])
    diagonais_iguais = (d[0] == d[1])
    lados_opstos_iguais = (l[0] == l[2] and l[1] == l[3])

    if lados_iguais and diagonais_iguais:
        print('Quadrado')
    elif diagonais_iguais and lados_opstos_iguais:
        print('a figura é um retângulo')
    else:
      print('a figura é um quadrilátero')
    #classifica o quadrilátero como quadrado, retângulo ou geral


def sugestoes():
    #dá sugestões para o usuário

    print("\n" + "="*50)
    print("SUGESTÕES DE COORDENADAS PARA TESTES RÁPIDOS:")
    print("="*50)

    print("\n--- Para 3 Coordenadas (Triângulos) ---")

    # Teste de Equilátero
    print("Equilátero (Lado 10):")
    print("  Ponto A: X=0, Y=0")
    print("  Ponto B: X=10, Y=0")
    print("  Ponto C: X=5, Y=8.660254")

    # Teste de Isósceles (Lados 5, 5, 6)
    print("\nIsósceles:")
    print("  Ponto A: X=0, Y=0")
    print("  Ponto B: X=6, Y=0")
    print("  Ponto C: X=3, Y=4")

    # Teste de Escaleno (Lados 5, 12, 13 - Triângulo Retângulo)
    print("\nEscaleno (Todos os lados diferentes):")
    print("  Ponto A: X=0, Y=0")
    print("  Ponto B: X=5, Y=0")
    print("  Ponto C: X=0, Y=12")

    print("\n--- Para 4 Coordenadas (Quadriláteros) ---")

    # Teste de Quadrado (Lado 5)
    print("Quadrado:")
    print("  Ponto A: X=0, Y=0")
    print("  Ponto B: X=5, Y=0")
    print("  Ponto C: X=5, Y=5")
    print("  Ponto D: X=0, Y=5")

    # Teste de Retângulo (Lados 4 e 6)
    print("\nRetângulo:")
    print("  Ponto A: X=0, Y=0")
    print("  Ponto B: X=6, Y=0")
    print("  Ponto C: X=6, Y=4")
    print("  Ponto D: X=0, Y=4")

    print("="*50 + "\n")


def principal():
    sugestoes()


    while True:
        escolha = input("Deseja inserir 3 ou 4 coordenadas? (Digite 3 ou 4): ")
        if escolha in ('3', '4'):
            num_pontos = int(escolha)
            break
        else:
            print("Escolha inválida. Por favor, digite '3' ou '4'.")

    #Coordenadas

    if num_pontos == 3:
        # Coleta 3 pontos (P1, P2, P3)
        p1 = coordenada(1, 'Ponto A')
        p2 = coordenada(2, 'Ponto B')
        p3 = coordenada(3, 'Ponto C')

        # Calcula os 3 lados
        lado_a = calcular_distancia(p1, p2)
        lado_b = calcular_distancia(p2, p3)
        lado_c = calcular_distancia(p3, p1)

        # Classifica o triângulo
        triangulo(lado_a, lado_b, lado_c)

    elif num_pontos == 4:
        # Coleta 4 pontos (P1, P2, P3, P4)
        p1 = coordenada(1, 'Ponto A')
        p2 = coordenada(2, 'Ponto B')
        p3 = coordenada(3, 'Ponto C')
        p4 = coordenada(4, 'Ponto D')

        pontos = [p1, p2, p3, p4]

        # Classifica o quadrilátero
        quadrilatero(pontos)

if __name__ == "__main__":
    principal()
```
